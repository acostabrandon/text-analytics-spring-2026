Step 3: Text Cleaning and Preprocessing

Because VADER uses punctuation/caps as signal, I will preserve a separate cleaning version called text_vader. For the full standard cleaning, a version called text_clean will be used for other models. 

```python
df["text_raw"] = df["ReviewBody"].fillna("").astype(str)
```

```python
#remove html and numbers
import re

df["text_no_html"] = df["text_raw"].str.replace(r"<.*?>", " ", regex=True)
```

```python
#remove urls
df["text_no_urls"] = df["text_no_html"].str.replace(r"http\S+|www\.\S+", " ", regex=True)
```

```python
#remove contractions using library
!pip install -q contractions
import contractions

df["text_expanded"] = df["text_no_urls"].apply(contractions.fix)
```

```python
#create vader version before addressing capitilization and punctuation
df["text_vader"] = df["text_expanded"].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
#create standard cleaned version by recoving whitespace, and making all lowercase
df["text_clean"] = df["text_expanded"].str.lower()

# remove non-letters (keep spaces)
df["text_clean"] = df["text_clean"].str.replace(r"[^a-z\s]", " ", regex=True)

# remove extra whitespace
df["text_clean"] = df["text_clean"].str.replace(r"\s+", " ", regex=True).str.strip()
```

Step 4 & 5: Tokenize, Feature Engineer, and Vectorize

Starting with VADER pipeline: 

```python
!pip install vaderSentiment

import vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
def vader_score(text):
    return sia.polarity_scores(str(text))  # returns neg/neu/pos/compound

def vader_label(compound):
    if compound >= 0.05:
        return "positive"
    elif compound <= -0.05:
        return "negative"
    else:
        return "neutral"

import time

start = time.time()

# --- your existing VADER code ---
df['vader_score'] = df['text_vader'].apply(vader_score)

vader_df = df["vader_score"].apply(pd.Series)

df["compound_score"] = vader_df["compound"]
df["positive_score"] = vader_df["pos"]
df["neutral_score"] = vader_df["neu"]
df["negative_score"] = vader_df["neg"]

df["vader_pred"] = df["compound_score"].apply(vader_label)
# --- end VADER code ---

vader_time = time.time() - start
print("VADER total time:", round(vader_time, 3), "sec")
print("Rows processed:", len(df))
print("Time per row:", round(vader_time / len(df), 6), "sec")
```

Vader tokenizes through simple word-level processing with punctuation handling. It uses a broad array of features, like negation, capitlization empphasis, punctuation intensity, and degree modifiers like "very". VADER expects minimal cleaning, since capitalization and punctuations can be preserved. 

TextBlob Pipeline: 

```python
#!pip install -q textblob
from textblob import TextBlob
```

```python
from textblob import TextBlob

import time

start = time.time()

# --- your existing TextBlob code ---
def textblob_scores(text):
    s = TextBlob(str(text)).sentiment
    return pd.Series([s.polarity, s.subjectivity])

tb_df = df["text_clean"].apply(textblob_scores)
tb_df.columns = ["tb_polarity", "tb_subjectivity"]

df["tb_polarity"] = tb_df["tb_polarity"]
df["tb_subjectivity"] = tb_df["tb_subjectivity"]

def textblob_label(p):
    if p > 0.05:
        return "positive"
    elif p < -0.05:
        return "negative"
    else:
        return "neutral"

df["tb_pred"] = df["tb_polarity"].apply(textblob_label)
# --- end TextBlob code ---

tb_time = time.time() - start
print("TextBlob total time:", round(tb_time, 3), "sec")
print("Rows processed:", len(df))
print("Time per row:", round(tb_time / len(df), 6), "sec")
```

TextBlob tokenizes with word-level tokenization. It scores the polarity/subectivity of each word to determine its sentiment. Standard cleaning is expected for this model, with lowercasing and removal of extra punctuations needed. 

Transformer Pipeline:

```python
import torch
import torch.nn.functional as F
import time
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("bhadresh-savani/distilbert-base-uncased-emotion")
model = AutoModelForSequenceClassification.from_pretrained("bhadresh-savani/distilbert-base-uncased-emotion")
model.eval()

labels = ["sadness", "joy", "love", "anger", "fear", "surprise"]

def get_emotion_scores(text):
    inputs = tokenizer(
        str(text),
        return_tensors="pt",
        truncation=True,
        max_length=512
    )
    with torch.no_grad():
        outputs = model(**inputs)
    probs = F.softmax(outputs.logits, dim=1).squeeze().tolist()
    return dict(zip(labels, probs))

start = time.time()

emotion_scores = df["text_clean"].head(200).apply(get_emotion_scores)

emotion_df = emotion_scores.apply(pd.Series)
df.loc[df.index[:200], labels] = emotion_df.values

# add prediction (highest emotion score)
df.loc[df.index[:200], "transformer_pred"] = df.loc[df.index[:200], labels].idxmax(axis=1)

tr_time = time.time() - start
print("Transformer total time:", round(tr_time, 3), "sec")
print("Rows processed:", 200)
print("Time per row:", round(tr_time / 200, 6), "sec")

df[["text_clean"] + labels + ["transformer_pred"]].head(10)
```

Step 6: Analyze and Review Results

We will extract 100 sample rows from the dataset, run the model on each one, and then extract as a csv file where we can manually input the real_y value. 

```python
n = 100
sample_df = df[["text_raw"]].sample(n, random_state=42).copy()
sample_df = sample_df.reset_index(drop=True)

sample_df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

sia = SentimentIntensityAnalyzer()

def vader_scores(text):
    s = sia.polarity_scores(str(text))  # dict: neg, neu, pos, compound
    return pd.Series([s["neg"], s["neu"], s["pos"], s["compound"]])

# add score columns
v_out = sample_df["text_raw"].apply(vader_scores)
v_out.columns = ["vader_neg", "vader_neu", "vader_pos", "vader_compound"]
sample_df = pd.concat([sample_df, v_out], axis=1)

# label from compound
def vader_label(comp):
    if comp >= 0.05:
        return "positive"
    elif comp <= -0.05:
        return "negative"
    else:
        return "neutral"

sample_df["vader_pred"] = sample_df["vader_compound"].apply(vader_label)

sample_df[["text_raw", "vader_neg", "vader_neu", "vader_pos", "vader_compound", "vader_pred"]].head()
```

```python
from textblob import TextBlob
import pandas as pd

def tb_scores(text):
    s = TextBlob(str(text)).sentiment
    return pd.Series([s.polarity, s.subjectivity])

tb_out = sample_df["text_raw"].apply(tb_scores)
tb_out.columns = ["tb_polarity", "tb_subjectivity"]
sample_df = pd.concat([sample_df, tb_out], axis=1)

def tb_label(p):
    if p > 0.05:
        return "positive"
    elif p < -0.05:
        return "negative"
    else:
        return "neutral"

sample_df["tb_pred"] = sample_df["tb_polarity"].apply(tb_label)

sample_df[["text_raw", "tb_polarity", "tb_subjectivity", "tb_pred"]].head()
```

```python
import torch
import torch.nn.functional as F
import pandas as pd

labels = ["sadness", "joy", "love", "anger", "fear", "surprise"]

def tr_probs(text):
    inputs = tokenizer(str(text), return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        logits = model(**inputs).logits
    probs = F.softmax(logits, dim=1).squeeze().tolist()
    return pd.Series(probs)

# add 6 probability columns
tr_out = sample_df["text_raw"].apply(tr_probs)
tr_out.columns = labels
sample_df = pd.concat([sample_df, tr_out], axis=1)

# top emotion + sentiment mapping
sample_df["tr_top_emotion"] = sample_df[labels].idxmax(axis=1)

def emotion_to_sentiment(e):
    if e in ["sadness", "anger", "fear"]:
        return "negative"
    elif e in ["joy", "love"]:
        return "positive"
    else:
        return "neutral"  # surprise

sample_df["tr_pred"] = sample_df["tr_top_emotion"].apply(emotion_to_sentiment)

sample_df[["text_raw"] + labels + ["tr_top_emotion", "tr_pred"]].head()
```

Now that we have extracted a csv file with empty real_y values, we will manually impute the ground truth sentiments to form a table that we can use to calculate the accuracy. 

```python
from sklearn.metrics import confusion_matrix

# load your filled file (rename if needed)
df_filled = pd.read_csv("filled_random_sample_review.csv")

# clean labels
for col in ["real_y", "vader_pred", "tb_pred", "tr_pred"]:
    df_filled[col] = df_filled[col].astype(str).str.strip().str.lower()

# keep only valid labeled rows
valid = df_filled["real_y"].isin(["positive", "neutral", "negative"])
df_filled = df_filled[valid].copy()

labels = ["negative", "neutral", "positive"]

models = ["vader_pred", "tb_pred", "tr_pred"]

for m in models:
    acc = (df_filled[m] == df_filled["real_y"]).mean()
    print("\n", m, "accuracy:", round(acc, 3))

    cm = confusion_matrix(df_filled["real_y"], df_filled[m], labels=labels)
    print(pd.DataFrame(cm, index=labels, columns=labels))
```
